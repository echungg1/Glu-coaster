```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data
bio_df = pd.read_csv("data/bio.csv")

# Create figure and axis
plt.figure(figsize=(10, 6))

# Create histogram
plt.hist(bio_df['BMI'], bins=30, edgecolor='black', alpha=0.7, color='skyblue')

# Customize the plot
plt.title('Distribution of BMI Values', fontsize=14, pad=15)
plt.xlabel('BMI (kg/m²)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.grid(True, alpha=0.3)

# Add BMI category reference lines
bmi_categories = [(18.5, 'Underweight'), (24.9, 'Normal'), (29.9, 'Overweight'), (float('inf'), 'Obese')]
colors = ['red', 'green', 'orange', 'red']
max_height = plt.ylim()[1]

for i, (threshold, label) in enumerate(bmi_categories[:-1]):
    plt.axvline(x=threshold, color=colors[i], linestyle='--', alpha=0.5)
    plt.text(threshold + 0.5, max_height * 0.95, label, 
             rotation=90, verticalalignment='top')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

# Read the data
cgm_df = pd.read_csv("data/CGMacros-001/CGMacros-001.csv")

# Convert timestamp to datetime
cgm_df['Timestamp'] = pd.to_datetime(cgm_df['Timestamp'])

# Create figure and axis
plt.figure(figsize=(12, 6))

# Plot both CGM readings
plt.plot(cgm_df['Timestamp'], cgm_df['Libre GL'], label='Libre', color='blue', alpha=0.7)
plt.plot(cgm_df['Timestamp'], cgm_df['Dexcom GL'], label='Dexcom', color='red', alpha=0.7)

# Customize the plot
plt.title('Continuous Glucose Monitoring: Libre vs Dexcom for Participant 1', fontsize=14, pad=15)
plt.xlabel('Time', fontsize=12)
plt.ylabel('Glucose Level (mg/dL)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend()

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Adjust layout to prevent label cutoff
plt.tight_layout()
plt.show()
```
